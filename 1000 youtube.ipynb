```python
# Brand And Collaborations
def analysis_brand_collaboration(df):
# streamers with high performance metrics receive more brand collaborations and marketing campaigns 
    plt.figure(figsize=(14,8))
    #Scatter plot Subscribers Vs Visits
    plt.subplot(2,2,1)
    sns.scatterplot(x='Suscribers',y='Visits',data=df, color='purple')
    plt.title("Market Campaign vs Visits")
    plt.xlabel('Subscribers')
    plt.ylabel('Visits')
    #Scatter plot Subscribers VS Links
    plt.subplot(2,2,2)
    sns.scatterplot(x='Suscribers',y='Links',data=df, color='Orange')
    plt.title("Market Campaign vs Links")
    plt.xlabel('Subscribers')
    plt.ylabel('Links')
    #Scatter plot Subscriber VS Comments
    plt.subplot(2,2,3)
    sns.scatterplot(x='Suscribers',y='Comments',data=df, color='red')
    plt.title("Market Campaign vs Comments")
    plt.xlabel('Subscribers')
    plt.ylabel('Comments')
    plt.tight_layout()
    plt.show()
    
if __name__ =="__main__":
    analysis_brand_collaboration(df)
```

# BenchMarking

```python
# calculate average performance metrics
average_metrics=df[['Suscribers','Visits','Likes','Comments']].mean()
```

```python
average_metrics
```

```python
#filter Streamer  with above average performane
top_performers=df[(df['Suscribers']>average_metrics['Suscribers']) &(df['Visits']>average_metrics['Visits'])&(df['Comments']>average_metrics['Comments'])&(df['Likes']>average_metrics['Likes'])]
```

```python
#display information about top performing content  creator
top_performers_info=top_performers[['Username','Suscribers','Likes','Comments']]
print('Top performing content creator')
print(top_performers_info)
```
